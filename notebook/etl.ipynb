```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ› ï¸  CONFIGURACIÃ“N - SOLO CAMBIA ESTO
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
csv_filename = 'coffee_sales.csv'  # â† CAMBIA AQUÃ el nombre de tu archivo CSV
print(f"ğŸ“ ConfiguraciÃ³n: Se cargarÃ¡ '{csv_filename}'")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”§ LÃ“GICA AUTOMÃTICA - NO MANIPULAR
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
import pandas as pd
import os

# Detectar automÃ¡ticamente el directorio del proyecto
current_dir = os.getcwd()
project_root = current_dir
while project_root != os.path.dirname(project_root):
    if os.path.exists(os.path.join(project_root, 'data')):
        break
    project_root = os.path.dirname(project_root)

data_path = os.path.join(project_root, 'data', csv_filename)
print(f"âœ… Ruta detectada automÃ¡ticamente: {data_path}")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“Š CARGA Y EXPLORACIÃ“N DE DATOS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
if os.path.exists(data_path):
    df = pd.read_csv(data_path)
    print(f"âœ… Â¡Datos cargados! {df.shape[0]} filas Ã— {df.shape[1]} columnas")
    df.head()
else:
    print(f"âŒ Archivo no encontrado: {data_path}")
    print("ğŸ“‚ Archivos CSV disponibles:")
    for file in os.listdir(os.path.join(project_root, 'data')):
        if file.endswith('.csv'):
            print(f"  ğŸ“„ {file}")
```

```python
# Mostrar solo las primeras filas
df.head()
```

```python
# Enriquecimiento de fechas (tu mejora clave)
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["quarter"] = df["date"].dt.quarter    # âœ¨ NUEVA: Trimestre
df["month"] = df["date"].dt.month        # âœ¨ NUEVA: Mes  
df["year"] = df["date"].dt.year          # âœ¨ NUEVA: AÃ±o
```

```python
# Mostrar solo las primeras filas
df.head()
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ” INFORMACIÃ“N DETALLADA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
print("ğŸ“Š INFORMACIÃ“N TÃ‰CNICA")
print("=" * 50)

print(f"ğŸ—‚ï¸ COLUMNAS ({len(df.columns)}):")
for i, col in enumerate(df.columns, 1):
    print(f"  {i:2d}. {col}")

print(f"\nğŸ“‹ TIPOS DE DATOS:")
df.info()

print(f"\nğŸ“ˆ ESTADÃSTICAS DESCRIPTIVAS:")
df.describe()
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ” CALIDAD DE DATOS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
print("ğŸ” ANÃLISIS DE CALIDAD DE DATOS")
print("=" * 50)

print("âŒ VALORES FALTANTES:")
valores_nulos = df.isnull().sum()
valores_nulos = valores_nulos[valores_nulos > 0].sort_values(ascending=False)
if len(valores_nulos) > 0:
    for col, count in valores_nulos.items():
        porcentaje = (count / len(df)) * 100
        print(f"  {col}: {count} ({porcentaje:.1f}%)")
else:
    print("  âœ… No hay valores faltantes")

print(f"\nğŸ”„ VALORES DUPLICADOS:")
print(f"  Filas duplicadas: {df.duplicated().sum()}")

print(f"\nğŸ“ DIMENSIONES FINALES:")
print(f"  Filas: {df.shape[0]:,}")
print(f"  Columnas: {df.shape[1]}")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# â˜• TIPOS DE CAFÃ‰ MÃS VENDIDOS - AÃ‘O 2024
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
import matplotlib.pyplot as plt

# âœ… FILTRAR POR AÃ‘O 2024 ANTES DE CALCULAR
ventas_cafe_2024 = df[df['year'] == 2024]['coffee_name'].value_counts().head(10)

plt.figure(figsize=(12, 8))
colors = ['#8B4513', '#A0522D', '#CD853F', '#DEB887', '#F4A460', '#D2691E', '#B8860B', '#996515', '#8B4513', '#A0522D']
bars = plt.bar(range(len(ventas_cafe_2024)), ventas_cafe_2024.values, color=colors)

plt.title('â˜• Top 10 Tipos de CafÃ© MÃ¡s Vendidos - AÃ±o 2024', fontsize=16, fontweight='bold', color='#2F1B14')
plt.xlabel('Tipo de CafÃ©', fontsize=12)
plt.ylabel('Cantidad Vendida', fontsize=12)
plt.xticks(range(len(ventas_cafe_2024)), ventas_cafe_2024.index, rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')

# Agregar valores en las barras
for bar, valor in zip(bars, ventas_cafe_2024.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1, str(valor), 
             ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("â˜• TOP 10 CAFÃ‰S MÃS VENDIDOS EN 2024:")
for cafe, cantidad in ventas_cafe_2024.items():
    print(f"  {cafe}: {cantidad} ventas")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# â˜• VENTAS DE CAFÃ‰ POR MES 2024 (LÃNEAS)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
import matplotlib.pyplot as plt

# âœ… FILTRAR POR AÃ‘O 2024 ANTES DE CALCULAR
ventas_mes = df[df['year'] == 2024].groupby('month')['price'].sum().sort_index()

# Crear el grÃ¡fico de LÃNEAS (NO BARRAS)
plt.figure(figsize=(12, 6))
plt.plot(ventas_mes.index, ventas_mes.values, marker='o', linewidth=3, markersize=8, color='#8B4513')

plt.title('â˜• Ingresos Totales de CafÃ© por Mes - AÃ±o 2024 (Pipeline ETL)', fontsize=16, fontweight='bold', color='#2F1B14')
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Ingresos Totales ($)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout()
plt.show()

print("ğŸ“Š INGRESOS POR MES - AÃ‘O 2024:")
for mes, ingresos in ventas_mes.items():
    meses = ['', 'Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
    print(f"  {meses[mes]} 2024: ${ingresos:,.2f}")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ“ˆ VENTAS POR TRIMESTRE - AÃ‘O 2024
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
import matplotlib.pyplot as plt

# âœ… FILTRAR POR AÃ‘O 2024 ANTES DE CALCULAR
ventas_trimestre = df[df['year'] == 2024].groupby('quarter')['price'].sum()

plt.figure(figsize=(10, 6))
bars = plt.bar(ventas_trimestre.index, ventas_trimestre.values, 
               color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'], 
               edgecolor='black', linewidth=1)

plt.title('ğŸ“ˆ Ingresos por Trimestre - AÃ±o 2024 (Pipeline ETL)', fontsize=16, fontweight='bold')
plt.xlabel('Trimestre', fontsize=12)
plt.ylabel('Ingresos Totales ($)', fontsize=12)
plt.xticks([1, 2, 3, 4], ['Q1 (Ene-Mar)', 'Q2 (Abr-Jun)', 'Q3 (Jul-Sep)', 'Q4 (Oct-Dic)'])
plt.grid(True, alpha=0.3, axis='y')

# Agregar valores en las barras
for bar, valor in zip(bars, ventas_trimestre.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50, 
             f'${valor:,.0f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("ğŸ“ˆ INGRESOS POR TRIMESTRE - 2024:")
for trimestre, ingresos in ventas_trimestre.items():
    trimestre_nombres = {1: 'Q1 (Ene-Mar)', 2: 'Q2 (Abr-Jun)', 3: 'Q3 (Jul-Sep)', 4: 'Q4 (Oct-Dic)'}
    print(f"  {trimestre_nombres[trimestre]}: ${ingresos:,.2f}")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ’° DISTRIBUCIÃ“N DE PRECIOS - AÃ‘O 2024
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
import matplotlib.pyplot as plt

# âœ… FILTRAR POR AÃ‘O 2024 ANTES DE CALCULAR
precios_2024 = df[df['year'] == 2024]['price']

plt.figure(figsize=(12, 6))
plt.hist(precios_2024, bins=20, edgecolor='black', alpha=0.7, color='#8B4513')

plt.title('ğŸ’° DistribuciÃ³n de Precios de CafÃ© - AÃ±o 2024', fontsize=16, fontweight='bold', color='#2F1B14')
plt.xlabel('Precio ($)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(True, alpha=0.3, axis='y')
plt.ticklabel_format(style='plain', axis='x')

plt.tight_layout()
plt.show()

print("ğŸ’° ESTADÃSTICAS DE PRECIOS - 2024:")
print(f"  Precio mÃ­nimo: ${precios_2024.min():.2f}")
print(f"  Precio mÃ¡ximo: ${precios_2024.max():.2f}")
print(f"  Precio promedio: ${precios_2024.mean():.2f}")
print(f"  Precio mediano: ${precios_2024.median():.2f}")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ• PATRONES DE VENTA POR HORA - AÃ‘O 2024
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
import matplotlib.pyplot as plt

# âœ… FILTRAR POR AÃ‘O 2024 ANTES DE CALCULAR
df_2024 = df[df['year'] == 2024].copy()
df_2024['hour'] = pd.to_datetime(df_2024['time'], format='%H:%M').dt.hour
ventas_hora = df_2024.groupby('hour')['price'].sum()

plt.figure(figsize=(14, 6))
plt.plot(ventas_hora.index, ventas_hora.values, marker='o', linewidth=2, markersize=6, color='#2F1B14')
plt.fill_between(ventas_hora.index, ventas_hora.values, alpha=0.3, color='#8B4513')

plt.title('ğŸ• Patrones de Venta por Hora del DÃ­a - AÃ±o 2024', fontsize=16, fontweight='bold', color='#2F1B14')
plt.xlabel('Hora del DÃ­a', fontsize=12)
plt.ylabel('Ingresos Totales ($)', fontsize=12)
plt.xticks(range(0, 24, 2))
plt.grid(True, alpha=0.3)
plt.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
plt.show()

print("ğŸ• PICO DE VENTAS POR HORA - 2024:")
hora_pico = ventas_hora.idxmax()
ingresos_pico = ventas_hora.max()
print(f"  Hora pico: {hora_pico:02d}:00 (${ingresos_pico:.2f})")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ğŸ”¥ HEATMAP TEMPORAL - MES VS HORA 2024
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
import matplotlib.pyplot as plt
import numpy as np

# âœ… FILTRAR POR AÃ‘O 2024 ANTES DE CALCULAR
df_2024 = df[df['year'] == 2024].copy()
df_2024['hour'] = pd.to_datetime(df_2024['time'], format='%H:%M').dt.hour

# Crear matriz de datos para heatmap
heatmap_data = df_2024.groupby(['month', 'hour'])['price'].sum().unstack(fill_value=0)

plt.figure(figsize=(15, 8))
plt.imshow(heatmap_data.values, cmap='YlOrRd', aspect='auto')
plt.colorbar(label='Ingresos Totales ($)')

plt.title('ğŸ”¥ Heatmap Temporal: Ingresos por Mes y Hora - AÃ±o 2024', fontsize=16, fontweight='bold')
plt.xlabel('Hora del DÃ­a', fontsize=12)
plt.ylabel('Mes', fontsize=12)
plt.xticks(range(0, 24, 2), [f'{h:02d}:00' for h in range(0, 24, 2)])
plt.yticks(range(12), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 
                       'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])

plt.tight_layout()
plt.show()
```
